```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from ipywidgets import interact


@interact(x=(0,815,10), W=[1350,1350/2,-135], A=[13.3,17.5,22.5,28.8,35.2,41.8],b=[9.5,12.5,16,19], sigma_stat = False, L_stat = False,Lb_stat = False )
def plot(x=325,W=1350,A=0.7,b=9.5, sigma_stat=False, L_stat=False,Lb_stat = False ):
  th = np.arctan(255/(815-2*x))
  p = np.arctan(255/(x))
 
  

  def lin(coord1,coord2, T=0):

    x = np.array([coord1[0],coord2[0]])
    y = np.array([coord1[1],coord2[1]])
    plt.plot(x,y, "o", color = "black")
    plt.plot(x,y, color = "black")
    length = np.sqrt((x[0]-x[1])**2+(y[0]-y[1])**2)
    T = round(T,1)
    sigma = round(T/A,1)
    if L_stat == True:
       plt.annotate(f'L={length.round(1)}mm', xy=(np.average(x),np.average(y)), color = "black")
    
    if Lb_stat == True:
      plt.annotate(f'L/b={round(length/b,1)}', xy=(np.average(x),np.average(y)-5), color = "darkgreen")
   
    

    if T <0:
      
      plt.annotate(f'T={T}N', xy=(np.average(x),np.average(y)+5), color = "red", fontweight = "bold")
      if sigma_stat == True:
        plt.annotate(f'$\sigma_C={sigma}N/mm^2$', xy=(np.average(x),np.average(y)+15), color = "red")
    elif T==0:
      plt.annotate(f'T={T}N', xy=(np.average(x),np.average(y)+5), color = "green", fontweight = "bold")
    else:
         
      plt.annotate(f'T={T}N', xy=(np.average(x),np.average(y)+5), color = "blue", fontweight = "bold")
      if sigma_stat == True:
        plt.annotate(f'$\sigma_T={sigma}N/mm^2$', xy=(np.average(x),np.average(y)+15), color = "blue")


  


  lin([0,0],[0,255])
  lin([0,0],[x,0],-1*W/np.tan(th))
  lin([x,0],[2*x,0],-1*W/np.tan(th))
  lin([2*x,0],[815,0],-1*W/np.tan(th))
  lin([815,0],[2*x,255],W/np.sin(th))
  lin([x,255],[2*x,255],W*(1/np.tan(p)+1/np.tan(th)))
  lin([0,255],[x,255],W*(1/np.tan(p)+1/np.tan(th)))
  lin([x,255],[x,0])
  lin([x,0],[2*x,255],-1*W/np.tan(p))
  lin([x,0],[0,255],1*W/np.tan(p))
  lin([2*x,0],[2*x,255])

    
  plt.rcParams['figure.figsize'] = [20, 10]
  plt.rcParams['figure.dpi'] = 200
  


  plt.show()
```
