```
import numpy as np
import matplotlib.pyplot as plt
```

```
from ipywidgets import interact


@interact(x=(0,815,10), W=(-675,1350,5), A=(0.7,1.1,0.1) )
def plot(x=300,W=1350,A=0.7):
  th = np.arctan(255/(815-2*x))
  p = np.arctan(255/(x))
 
  

  def lin(coord1,coord2, T=0):

    x = np.array([coord1[0],coord2[0]])
    y = np.array([coord1[1],coord2[1]])
    plt.plot(x,y, "o", color = "black")
    plt.plot(x,y, color = "black")
    length = np.sqrt((x[0]-x[1])**2+(y[0]-y[1])**2)
    T = round(T,1)
    sigma = round(T/A,1)

    plt.annotate(f'T={T}N', xy=(np.average(x),np.average(y)+5), color = "blue")
    plt.annotate(f'L={length.round(1)}mm', xy=(np.average(x),np.average(y)), color = "red")
    if T <0:

      plt.annotate(f'$\sigma_c={sigma}N/mm^2$', xy=(np.average(x),np.average(y)+15), color = "magenta")
    
  # plt.annotate(f'{[x[0],y[0]]}', xy=(x[0],y[0]), color = "black")

 # plt.annotate(f'{[x[1],y[1]]}', xy=(x[1],y[1]), color = "black")

  


  lin([0,0],[0,255])
  lin([0,0],[x,0],-1*W/np.tan(th))
  lin([x,0],[2*x,0],-1*W/np.tan(th))
  lin([2*x,0],[815,0],-1*W/np.tan(th))
  lin([815,0],[2*x,255],W/np.sin(th))
  lin([x,255],[2*x,255],W*(1/np.tan(p)+1/np.tan(th)))
  lin([0,255],[x,255],W*(1/np.tan(p)+1/np.tan(th)))
  lin([x,255],[x,0])
  lin([x,0],[2*x,255],-1*W/np.tan(p))
  lin([x,0],[0,255],1*W/np.tan(p))
  lin([2*x,0],[2*x,255])





  plt.show()
```
